```r
res1 <- read.csv("result1.csv", header = FALSE)
res2 <- read.csv("result2.csv", header = FALSE)
res4 <- read.csv("result3.csv", header = FALSE)
res5 <- read.csv("result4.csv", header = FALSE)

names(res1)[1] <- "MCS"
names(res1)[2] <- "D"

names(res2)[1] <- "MCS"
names(res2)[2] <- "D"

names(res4)[1] <- "MCS"
names(res4)[2] <- "D"

names(res5)[1] <- "MCS"
names(res5)[2] <- "D"

c <- res1[1,2]
res1 <- res1[-c(1),]
res1$C <- c

c <- res2[1,2]
res2 <- res2[-c(1),]
res2$C <- c

tail(res2, 6)

c <- res4[1,2]
res4 <- res4[-c(1),]
res4$C <- c

c <- res5[1,2]
res5 <- res5[-c(1),]
res5$C <- c

res3 <- rbind(res1, res2, res4, res5) #, res4, res5)

head(res3, 5)
tail(res3, 5)
```
