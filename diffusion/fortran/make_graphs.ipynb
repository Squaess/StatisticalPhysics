```r
library(ggplot2)
```

```r
res1 <- read.csv("resul1.csv", header = FALSE)
res2 <- read.csv("resul2.csv", header = FALSE)
res4 <- read.csv("resul3.csv", header = FALSE)
res5 <- read.csv("resul4.csv", header = FALSE)

names(res1)[1] <- "MCS"
names(res1)[2] <- "D"

names(res2)[1] <- "MCS"
names(res2)[2] <- "D"

names(res4)[1] <- "MCS"
names(res4)[2] <- "D"

names(res5)[1] <- "MCS"
names(res5)[2] <- "D"

c <- res1[1,2]
res1 <- res1[-c(1),]
res1$C <- c

c <- res2[1,2]
res2 <- res2[-c(1),]
res2$C <- c

c <- res4[1,2]
res4 <- res4[-c(1),]
res4$C <- c

c <- res5[1,2]
res5 <- res5[-c(1),]
res5$C <- c

res3 <- rbind(res1, res2, res4, res5)

head(res3, 5)
tail(res3, 5)
```

```r
ggplot() + 
geom_point(data = res1, aes(MCS, D, color="red")) +
geom_point(data = res2, aes(MCS, D, color="blue")) +
geom_point(data = res4, aes(MCS, D, color="green")) +
geom_point(data = res5, aes(MCS, D, color="white")) +
ylim(0,0.5)
```

```r
res.v2 <- read.csv("results1.csv")
res.v2$C <- round(res.v2$C,1)
final.v2 <- read.csv("results2.csv")

# head(res.v2, 5)
# head(final.v2, 5)

head(res.v2, 5)
tail(res.v2,5)

tmp <- res.v2[round(res.v2$C,1),]
tail(tmp, 4)
```

```r
ggplot(data = res.v2) + 
geom_point(data = res.v2[res.v2$C == 0.1,], aes(MCS, D), colour = "blue") +
geom_point(data = res.v2[res.v2$C == 0.2,], aes(MCS, D), colour = "pink") +
geom_point(data = res.v2[res.v2$C == 0.3,], aes(MCS, D), colour = "red") +
geom_point(data = res.v2[res.v2$C == 0.4,], aes(MCS, D), colour = "green") + 
geom_line(data = res.v2[res.v2$C == 0.5,], aes(MCS, D), colour = "brown") + 
geom_point(data = res.v2[res.v2$C == 0.5,], aes(MCS, D), colour = "brown") + 
geom_point(data = res.v2[res.v2$C == 0.6,], aes(MCS, D), colour = "tan2") + 
geom_point(data = res.v2[res.v2$C == 0.7,], aes(MCS, D), colour = "orchid") +
geom_point(data = res.v2[res.v2$C == 0.8,], aes(MCS, D), colour = "snow4") + 
geom_point(data = res.v2[res.v2$C == 0.9,], aes(MCS, D), colour = "darkkhaki") + 
geom_point(data = res.v2[res.v2$C == 1.0,], aes(MCS, D), colour = "black") +
ylim(0,0.4)+
scale_fill_discrete(name="Dupa1\ndupa2",
                         labels=c("Control", "Treatment 1", "Treatment 2", "4", "5", "6", "7", "8", "9", "12")) + 
theme(legend.position = "bottom")
```
