```r
# ggplot() + geom_point(data = data, aes(T, S), color="red") + geom_point(data = data, aes(T, S2), color="blue") +
ggplot() + geom_point(data = data, aes(T, S2), color="blue") +
ylab("S") +
geom_hline(aes(yintercept=0.4, linetype = "0.4"), color="red") +
geom_vline(aes(xintercept=0.835, linetype = "0.835"), color="green") +
scale_linetype_manual(name = "line", values=c(2,2),
                      guide = guide_legend(override.aes = list(color = c("red", "green"))))
```
