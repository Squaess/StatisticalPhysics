```r
mag10 <- read_csv("mag10.csv")
mag50 <- read_csv("mag50.csv")
mag100 <- read_csv("mag100.csv")

ggplot() + geom_point(data = mag10, aes(T, m)) +
geom_point(data = mag50, aes(T, m), color="red") +
geom_point(data = mag100, aes(T, m), color="green")
```
