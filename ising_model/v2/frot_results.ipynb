```r
flips.3 <- read_csv("flips12.csv")

flips.3 <- flips.3[-c(1:100),]
flips.3 <- flips.3[abs(flips.3$m) > 0.7,]
flips.3 <- flips.3[flips.3$MCS %% 10000 == 0,]
ggplot() + geom_point(data = flips.3, aes(MCS, m), size = 3) + ylim(-1,1) + 
ggtitle("Magnetization (average spin) for T*=1.7 and L=10")
```

```r
mag10 <- read_csv("mag10.csv")
mag50 <- read_csv("mag50.csv")
mag100 <- read_csv("mag100.csv")
mag <- rbind(mag10, mag50, mag100)
ggplot(data = mag, aes(T, m, group=L, color=factor(L), shape=factor(L))) + 
geom_point() + geom_line() +
scale_color_discrete(name = "L") + 
scale_shape_discrete(name = "L") + 
ggtitle("Temperature dependence of the mean value of magnetization") + ylab("<m>") +
theme(legend.position="top")
# ggplot() + 
# geom_point(data = mag10, aes(T, m), shape = 15) + 
# geom_line(data = mag10, aes(T, m)) +
# geom_point(data = mag50, aes(T, m), color="red", shape = 16) +
# geom_line(data = mag50, aes(T, m), color="red") +
# geom_point(data = mag100, aes(T, m), color="green", shape = 17) +
# geom_line(data = mag100, aes(T, m), color="green") +
```

```r
energ10 <- read_csv("ener10.csv")
energ50 <- read_csv("ener50.csv")
energ100 <- read_csv("ener100.csv")

energ <- rbind(energ10, energ50, energ100)

ggplot(data = energ, aes(T, e, group=L, color=factor(L))) + 
geom_point(size = 1) + geom_line() + 
scale_color_discrete(name = "L") + 
scale_shape_discrete(name = "L") + 
ggtitle("Temperature dependence of the mean value of energy") +
theme(legend.position="right") + ylab("energy")
# geom_point(data = energ50, aes(T, e), color = "red") + 
# geom_point(data = energ100, aes(T, e), color = "green")
```

```r
bind10 <- read_csv("binder10.csv")
bind50 <- read_csv("binder50.csv")
bind100 <- read_csv("binder100.csv")

bind <- rbind(bind10, bind50, bind100)
bind <- bind[bind$T < 2.43 & bind$T > 1.9,]

ggplot(data = bind, aes(T, b, group=L, color=factor(L))) + geom_point() + geom_line() +
ggtitle("Temperature dependence of the mean value of Binder’s cumulant") +
scale_color_discrete(name = "L") + 
scale_shape_discrete(name = "L") + 
theme(legend.position="top") + ylab("U")
```
